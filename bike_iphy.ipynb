```python
import pylab
import calendar
import numpy as np
import pandas as pd
import seaborn as sn
from scipy import stats
import missingno as msno
from datetime import datetime
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
pd.options.mode.chained_assignment = None
warnings.filterwarnings("ignore", category=DeprecationWarning)
%matplotlib inline
#這邊是在import會用到的東西
```

```python
train = pd.read_csv('/content/bike/train.csv')
#這是我放在colab的路徑
```

```python
train.head()#可看到前五筆資料
```

```python
train[train.duplicated()] 
#前面都在看跟整理資料(資料預處理)，下面的順序是下面的順序是決策樹，RandomForest跟KNN
```

```python
train.drop_duplicates(inplace = True)
```

```python
train.corr()
```

```python
X = train[["season","workingday","temp","atemp","windspeed"]].values
X[0:3]
```

```python
Y = train[["count"]].values
Y[0:3]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_temp, Y_train, Y_temp = train_test_split(
    X, Y, train_size=0.6, random_state=0)

X_test, X_vali, Y_test, Y_vali = train_test_split(
    X_temp, Y_temp, test_size=0.5, random_state=0)
#這邊再切資料，切成6:2:2
```

```python
#決策樹
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier(criterion='entropy', max_depth=13)

model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

print('Decision Tree')
print('Accuracy: ' + str(accuracy_score(Y_test, y_pred)))
#這邊是在用決策樹演算法跟算印出成功率
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100)

model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

print('Random Forest')
print('Accuracy: ' + str(accuracy_score(Y_test, y_pred)))
#這邊是在用RandomForest演算法跟印出成功率
```

```python
from sklearn.neighbors import KNeighborsRegressor
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
length = round(sqrt(train.shape[0]))
```

```python
rmse_dict = {}
rmse_list = []
for k in range(1,length+1):
    model = KNeighborsRegressor(n_neighbors = k).fit(X_train,Y_train)
    Y_predict = model.predict(X_test)
    rmse = sqrt(mean_squared_error(Y_test,Y_predict))
    rmse_dict.update({k:rmse})
    rmse_list.append(rmse)
    print("Rmse for k = {} is {}" .format(k,rmse))
    #這邊是KNN演算法的過程，我把它印出來
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=69)

model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

print('kNN')
print('Accuracy: ' + str(accuracy_score(Y_test, y_pred)))
#這邊是在印出KNN的成功率
```

```python
key_min = min(rmse_dict.keys(), key=(lambda k: rmse_dict[k]))

print( "The miminum RMSE value is ",rmse_dict[key_min], "with k= ", key_min) 
#這是KNN的最小值
```
